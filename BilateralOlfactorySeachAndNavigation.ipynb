#The Dynamics of Bilateral Olfactory Search and Navivation:
https://arxiv.org/pdf/2008.03358.pdf

```python
# Packages

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Setting up the system of ODEs

def system(y, t, beta, cL, cR):
    x, y, theta = y
    dx_dt = np.cos(theta)
    dy_dt = np.sin(theta)
    dtheta_dt = beta * (cL - cR) # For the simplest case we will choose constant concentrations cL and cR
    return [dx_dt, dy_dt, dtheta_dt]
```

```python
# Simulation parameters

T = 10  # simulation time
dt = 0.01  # time step
t = np.arange(0, T + dt, dt)  # time vector
N = len(t)  # number of time steps
```

```python
# Model parameters

beta = 0.5
cL = 3.0
cR = 2.0
y0 = [1, 0, 0]  # initial conditions
```

```python
# Solving the system of ODEs

sol = odeint(system, y0, t, args=(beta, cL, cR))
```

```python
# Plotting the results

# Plot the results
fig, ax = plt.subplots()
ax.plot(sol[:, 0], sol[:, 1])
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_title('Trajectory')
plt.show()

fig, ax = plt.subplots()
ax.plot(t, sol[:, 2])
ax.set_xlabel('Time')
ax.set_ylabel('theta(t)')
ax.set_title('Orientation')
plt.show()
```
